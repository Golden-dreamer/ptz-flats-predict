Данные, что содержатся в "Вид из окон" преобразовываем в one-hot encoding

```python
def get_unique_list(df, col):
    uniques = []
    for var in list(df[col].dropna().unique()):
        un = var.split(',')
        for value in un:
            if value.strip() not in uniques:
                uniques.append(value.strip())
    return uniques
```

```python
uniques = get_unique_list(df, 'Вид из окон')
```

```python
courtyard = pd.DataFrame(index=df.index,columns=uniques)
```

```python
def oh_transform(df, col, out_df):
    for i,v in df[col].dropna().items():
        vals = v.split(',')
        value = [val.strip() for val in vals]
        out_df.loc[i,value] = 1
    
```

```python
for i,v in df['Вид из окон'].dropna().items():
    vals = v.split(',')
    value = [val.strip() for val in vals]
    courtyard.loc[i,value] = 1
```
